```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer


df['genres_list'] = df['genres'].apply(lambda x: [g.strip() for g in x.split(",")])
mlb = MultiLabelBinarizer()
genre_one_hot = mlb.fit_transform(df['genres_list'])

print("All unique genres found:")
print(mlb.classes_)

genre_df = pd.DataFrame(genre_one_hot, columns=mlb.classes_)

df = pd.concat([df, genre_df], axis=1)

df.head()
```
